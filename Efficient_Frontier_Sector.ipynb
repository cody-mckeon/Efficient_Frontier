```python
# Portfolio Construction
# Simple equal weighting
# For more complex weighting: 
    # Market Cap Weighting
    # Volatility-Based Weighting (Risk Parity)
    # Factor-Based Weighting
    # Optimization-Based Weighting (Efficient Frontier) 

# Create Sector allocation weights based on sector size
def calculate_sector_weights(sector_info, selected_data, sector_allocation):
    stock_weights = {}
    
    # Calculate the weight of each stock within its sector
    sector_allocation_dict = sector_allocation.to_dict()
    for sector, sector_weight in sector_allocation_dict.items():
        sector_stocks = sector_info[sector_info['GICS Sector'] == sector]['Symbol'].tolist()
        stock_weight_per_stock = sector_weight / len(sector_stocks)
        
        for stock in sector_stocks:
            if stock in selected_data.columns:
                stock_weights[stock] = stock_weight_per_stock

    # Convert stock weights into a NumPy array
    weights = np.array([stock_weights.get(stock, 0) for stock in selected_data.columns])
    
    # Normalize the weights to ensure they sum to 1
    weights /= np.sum(weights)
    
    return weights
```

```python
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import objective_functions
from pypfopt import expected_returns, risk_models

# Optimize portfolio with Efficient Frontier
def optimize_portfolio(selected_data, sector_weights):
    # Calculate expected returns and the covariance matrix
    mu = expected_returns.mean_historical_return(selected_data)
    S = risk_models.sample_cov(selected_data)

    # Ensure the covariance matrix is symmetric
    S = (S + S.T) / 2
    
    # Initialize Efficient Frontier object
    ef = EfficientFrontier(mu, S)
    
    # Add sector weight constraints
    ef.add_objective(objective_functions.L2_reg, gamma=0.1)  # Optional L2 regularization for smoother weights

    # Set initial sector-based weights
    try:
        ef.min_volatility()
    except ValueError as e:
        print("Error during optimization:", e)
        return None, None
    
    # Get optimized weights
    cleaned_weights = ef.clean_weights()
    
    # Calculate portfolio performance
    performance = ef.portfolio_performance(verbose=True)
    
    return cleaned_weights, performance

# Fetch stock weights from sector allocation
sector_weights = calculate_sector_weights(remaining_info, selected_data_cleaned, sector_allocation)

# Optimize portfolio
weights, performance = optimize_portfolio(selected_data_cleaned, sector_weights)
print("Optimized Weights:", weights)

```

```python
# Rebalance weights to maintain sector allocation constraints
def rebalance_weights(cleaned_weights, sector_info, sector_allocation):
    rebalanced_weights = {}
    
    # Group stocks by sector and adjust their weights
    for sector, sector_weight in sector_allocation.items():
        sector_stocks = [stock for stock in cleaned_weights.keys() if stock in sector_info[sector_info['GICS Sector'] == sector]['Symbol'].values]
        
        # Adjust stock weights within the sector
        # Here I can put the desired sector allocation
        total_sector_weight = sum([cleaned_weights[stock] for stock in sector_stocks])
        for stock in sector_stocks:
            rebalanced_weights[stock] = cleaned_weights[stock] * (sector_weight / total_sector_weight)
    
    return rebalanced_weights

# Rebalance optimized weights based on sector constraints
rebalanced_weights = rebalance_weights(weights, remaining_info, sector_allocation)
print("Rebalanced Weights:", rebalanced_weights)
```

For each stock, the simple return between two time points is calculated as:

$$
R_t = \frac{P_t}{P_{t-1}} - 1
$$

Where:
- \( R_t \) is the return at time \( t \),
- \( P_t \) is the price at time \( t \),
- \( P_{t-1} \) is the price at time \( t-1 \).

This gives the percentage change in price.

This is done by multiplying the mean daily return by the number of trading periods in a year (usually 252 for daily returns in stock markets):

$$
\text{Annualized Return} = (1 + \text{Average Daily Return})^{252} - 1
$$

Or for simple annualization (which is what the function does by default):

$$
\text{Annualized Return} = \text{Average Daily Return} \times 252
$$

This gives an estimate of the return assuming the pattern of daily returns holds throughout the year.


While simple returns are intuitive, in many cases, logarithmic returns are preferred because they account for compounding over time and have better statistical properties (e.g., they are symmetric and additive over time).

To calculate logarithmic returns:

$$
\text{Logarithmic Return} = \ln\left(\frac{P_t}{P_{t-1}}\right)
$$
